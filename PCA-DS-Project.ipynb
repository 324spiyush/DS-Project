```python
from sklearn.model_selection import train_test_split 
import numpy as pmk 
from matplotlib.colors import ListedColormap 
from sklearn.linear_model import LogisticRegression 
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as biso 
from matplotlib.colors import ListedColormap 
from sklearn.metrics import confusion_matrix 
from sklearn.decomposition import PCA 
import pandas as khat 
```

```python
set_data = khat.read_csv('wines.csv')
y = set_data.iloc[:, 13].values  
X = set_data.iloc[:, 0:13].values 
shyam, kumar, piyush, pavithra = train_test_split(X, y, test_size = 0.2, random_state = 0)
sc = StandardScaler() 
shyam = sc.fit_transform(shyam) 
kumar = sc.transform(kumar) 
pca = PCA(n_components = 2) 
shyam = pca.fit_transform(shyam) 
kumar = pca.transform(kumar) 
explained_variance = pca.explained_variance_ratio_ 
classifier = LogisticRegression(random_state = 0) 
classifier.fit(shyam, piyush) 
y_pred = classifier.predict(kumar) 
cm = confusion_matrix(pavithra, y_pred) 
dhoom, verma = shyam, piyush 
```

```python
X1, X2 = pmk.meshgrid(pmk.arange(start = dhoom[:, 0].min() - 1, stop = dhoom[:, 0].max() + 1, step = 0.01), pmk.arange(start = dhoom[:, 1].min() - 1, stop = dhoom[:, 1].max() + 1, step = 0.01)) 
biso.contourf(X1, X2, classifier.predict(pmk.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape), alpha = 0.75, cmap = ListedColormap(('yellow', 'white', 'aquamarine'))) 
biso.xlim(X1.min(), X1.max()) 
biso.ylim(X2.min(), X2.max()) 
for i, j in enumerate(pmk.unique(verma)): 
	biso.scatter(dhoom[verma == j, 0], dhoom[verma == j, 1], c = ListedColormap(('red', 'green', 'blue'))(i), label = j) 
biso.title('Logistic Regression (Training set)') 
biso.xlabel('PC1')
biso.ylabel('PC2') 
biso.legend()
biso.show()  
dhoom, verma = kumar, pavithra 
```
